```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

# 1. Pre-Processing

```python
df = pd.read_csv("/Users/hannahwurzel/Desktop/dl-firefighters/DL-firefighters/data/data660k.csv")
df.head()
```

```python
df.columns
```

### 1a. Removing unnecessary columns

```python
data = df.drop(columns=["Unnamed: 0", "fire_ID", "start_date", "end_date", "direction_s", "landcover_s", "tile_ID", "geometry"])
data.head()
```

### 1b. Scale the data using Min/Max Normalization

```python
data_temp = data.drop('expansion', axis=1)
df_norm = (data_temp-data_temp.min())/(data_temp.max()-data_temp.min())
df_norm = pd.concat((df_norm, data.expansion), 1)
```

```python
df_norm.head()
```

### 1c. Split data in train and test sets

```python
train_df, test_df = train_test_split(df_norm, test_size=0.2)
```

```python
print("Length of train dataset: ", len(train_df), "\nLength of test dataset: ", len(test_df))
```
